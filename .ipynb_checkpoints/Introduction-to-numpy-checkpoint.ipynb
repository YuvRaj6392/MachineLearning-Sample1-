## Dot element example

```python
np.random.seed(7)
sales_amount=np.random.randint(20,size=(5,3))
sales_amount
```

```python
weekly_sales=pd.DataFrame(sales_amount,index=["Mon","Tue","Wed","Thu","Fri"],columns=["Almond","Peanut","Cashew"])
weekly_sales
```

```python
prices=np.array([20,40,80])
prices
```

```python
butter_prices=pd.DataFrame(prices.reshape(1,3),index=["Price"],columns=["Almond","Peanut","Cashew"])
butter_prices
```

```python
Total=np.dot(weekly_sales,butter_prices.T)
Total
```

```python
weekly_sales["Total"]=Total
```

```python
butter_prices
```

```python
weekly_sales
```
